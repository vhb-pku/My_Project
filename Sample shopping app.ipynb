```python
import pandas as pd
import numpy as np
df_des = pd.read_csv("df_des.csv")
df_perf = pd.read_csv("df_perf.csv")
df_query_perf = pd.read_csv("df_query_perf.csv")
```

### I.PART

```python
df_query_perf["conversion_rate"] = df_query_perf["orders"] / df_query_perf["hits"]
```

```python
df_query_perf.loc[df_query_perf.clicks == 0]  # lets drop this query, because it includes a wrong value.
```

```python
df_query_perf.drop(2, inplace = True)
```

```python
df_query_perf.reset_index(drop= True)
df_query_perf.head()
```

```python
df_query_perf["add_cart_to_exit_rate"] = (df_query_perf.carts- df_query_perf.orders) / (df_query_perf.carts)
df_query_perf.fillna(0, inplace = True)
```

```python
df_query_perf.head()
```

```python
df_query_perf.sort_values("conversion_rate", ascending = False, inplace = True)
```

```python
df_query_perf[:10]   # This 10 queries have really conversion rates. But are they reliable?
```

```python
df_query_perf.tail(10)  # These 10 unique queries have the lowest conversion rates comparing to the dataframe. 
# But same question applies again in here, are they reliable?
```

```python
df_query_perf.conversion_rate.mean()
```

```python
df_query_perf.loc[(df_query_perf.hits > 60)][:10]

#Without a doubt, these rows are more reliable than the others. 
#So Havana Dresses are basically the query that has most conversion rate! We should definitely focus more on these items!
```

```python
df_query_perf.loc[(df_query_perf.hits > 60)].tail(10)
#This queries have the lowest conversion rates. So we should focus on improving these queries to gain more conversions.
```

```python
df_query_perf.loc[df_query_perf.clicks > df_query_perf.hits + 100]

# Even though this rows have more clicks, their hit numbers are really low. We can say that probably their bounce rate is very high.
# Some users will probably just click 3-4 fields and leave it like that. 
# Our focus must be on improving the site and UX for reducing the bounce ratio.
```

```python
df_query_perf.sort_values("ctr", ascending = False)[:10]

# Basically dresses are taking the lead! ( VICTORIA Secret is 1st place ofcourse)
# But even though their CT rates are high, the add_cart_to_exit_rate is still too big. Therefore we can say that these users just want to check,
# because of the ad's attractiveness.
```

```python
df_perf.rename(columns = {'pId' : "PID"}, inplace =True)
```

```python
df = pd.merge(df_des, df_perf, on = "PID")
```

```python
df.drop(["Final Price", "Initial Price"], 1, inplace = True)
```

```python
df.head(5)
```

```python
df["category_order_mean"] = df.groupby("Category")["orders"].transform("mean")  #encoded
```

```python
df["category_clicks_mean"] = df.groupby("Category")["clicks"].transform("mean")
```

```python
df["cart_to_conversion"] = (df.carts - df.orders) / df.carts
```

```python
df["category_cart_to_conversion_mean"] = df.groupby("Category")["cart_to_conversion"].transform("mean")
```

```python
df.isna().any()
```

```python
df.loc[df.cart_to_conversion.isnull()]
```

```python
df.cart_to_conversion.fillna(0, inplace = True)
df.category_cart_to_conversion_mean.fillna(0, inplace = True)

# Because they dont even have any conversions!
```
